```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"E:\Raghav\scienccceee\science\data\bankloan\credit_train.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=["Months since last delinquent"],inplace=True)
```

```python
dataset.dropna(inplace=True)
```

```python
sns.displot(dataset["Credit Score"])
plt.show()
```

```python
q1 = dataset["Credit Score"].quantile(0.25)
q3 = dataset["Credit Score"].quantile(0.75)
iqr = q3-q1
```

```python
min_r = q1-(1.5*iqr)
max_r = q3+(1.5*iqr)
```

```python
min_r,max_r
```

```python
# dataset = dataset[dataset["Credit Score"]<=max_r]
```

```python
sns.distplot(dataset["Credit Score"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func= np.log1p)
```

```python
ft.fit([["Credit Score"]])
```

```python
dataset["Credit Score_tf"] =ft.transform(dataset[["Credit Score"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset["Credit Score"])
plt.title("before")

plt.subplot(1,2,2)
sns.displot(dataset["Credit Score_tf"])
plt.title("After")
plt.show()
```

```python
ft1 = FunctionTransformer(func= lambda x : x **2)
ft1.fit(dataset[["Credit Score"]])
dataset["Credit Score_tf1"] =ft1.transform(dataset[["Credit Score"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset["Credit Score"])
plt.title("before")

plt.subplot(1,2,2)
sns.displot(dataset["Credit Score_tf1"])
plt.title("After")
plt.show()
```

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"E:\Raghav\scienccceee\science\data\bankloan\credit_train.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=["Months since last delinquent"],inplace=True)
```

```python
dataset.dropna(inplace=True)
```

```python
sns.displot(dataset["Monthly Debt"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["Credit Score"]])
```

```python
dataset[["Credit Score"]] = ms.transform(dataset[["Credit Score"]])
```

```python
dataset.head(3)
```

```python
ss = StandardScaler()
ss.fit(dataset[["Credit Score"]])
```

```python
dataset["Credit Score_ss"] = pd.DataFrame(ss.transform(dataset[["Credit Score"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("before")
sns.displot(dataset["Credit Score"])
plt.subplot(1,2,2)
plt.title("After")
sns.displot(dataset["Credit Score_ss"])
plt.show()
```
